```python
# Weak vs Strong prompts

prompts = {
    "Resume": {
        "Weak": "Write resume",
        "Strong": "Create a professional resume for a computer science student with skills in Python, Java, and Machine Learning."
    },
    "Business Idea": {
        "Weak": "Give business idea",
        "Strong": "Suggest a low-budget AI-based startup idea suitable for college students in India."
    },
    "Study Plan": {
        "Weak": "Make study plan",
        "Strong": "Create a 7-day study plan for learning Machine Learning with theory and practical tasks."
    }
}

# Display prompts
for category, p in prompts.items():
    print(f"{category}:")
    print(f"Weak Prompt: {p['Weak']}")
    print(f"Strong Prompt: {p['Strong']}\n")
```
