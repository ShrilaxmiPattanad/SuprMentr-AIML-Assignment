```python
from nltk.corpus import wordnet as wn
import nltk

nltk.download('wordnet')

# Word pairs
pairs = [
    ("car", "automobile"),
    ("big", "large"),
    ("happy", "joyful"),
    ("teacher", "student"),
    ("dog", "animal")
]

# Calculate similarity
for w1, w2 in pairs:
    syn1 = wn.synsets(w1)
    syn2 = wn.synsets(w2)

    if syn1 and syn2:
        similarity = syn1[0].wup_similarity(syn2[0])
        print(f"{w1} - {w2}: Similarity = {similarity}")
```
