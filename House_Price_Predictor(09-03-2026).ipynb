```python
# House Price Predictor using Linear Regression

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step 1: Create dataset
data = {
    "Area_sqft": [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],
    "Bedrooms": [2, 2, 3, 3, 4, 4, 4, 5],
    "House_Age": [10, 8, 6, 5, 4, 3, 2, 1],
    "Price": [3000000, 3500000, 4200000, 5000000, 6200000, 6800000, 7200000, 8000000]
}

df = pd.DataFrame(data)

print("House Dataset:")
print(df)

# Step 2: Features and target
X = df[["Area_sqft", "Bedrooms", "House_Age"]]
y = df["Price"]

# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Step 4: Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 5: Predict new house price
new_house = pd.DataFrame({
    "Area_sqft": [1600],
    "Bedrooms": [3],
    "House_Age": [4]
})

predicted_price = model.predict(new_house)

print("\nPredicted House Price:", int(predicted_price[0]))
```
