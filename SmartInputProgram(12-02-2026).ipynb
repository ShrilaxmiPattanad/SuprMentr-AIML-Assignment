```python
# 🌟 Smart Input Program

print("✨ Welcome to the Smart Personality Builder ✨\n")

# Taking inputs
name = input("Enter your name: ").strip().title()

# Age validation
while True:
    try:
        age = int(input("Enter your age: "))
        if age <= 0:
            print("Age must be a positive number. Try again.")
        else:
            break
    except ValueError:
        print("Please enter a valid number for age.")

hobby = input("Enter your favorite hobby: ").strip().lower()

# Age categorization
if age <= 12:
    category = "Curious Explorer 🧸"
elif 13 <= age <= 19:
    category = "Energetic Teen ⚡"
elif 20 <= age <= 35:
    category = "Young Achiever 🚀"
elif 36 <= age <= 59:
    category = "Experienced Leader 🎯"
else:
    category = "Wise Legend 🌟"

# Hobby-based bonus message
if hobby in ["reading", "coding", "studying"]:
    hobby_message = "You love gaining knowledge! Keep sharpening your brain 🧠."
elif hobby in ["sports", "cricket", "football", "badminton"]:
    hobby_message = "You are full of energy and team spirit! 🏆"
elif hobby in ["music", "dance", "singing", "art"]:
    hobby_message = "You have a creative soul! 🎨🎵"
else:
    hobby_message = "That’s a wonderful hobby! Keep enjoying it! 😊"

# Final Personalized Output
print("\n" + "="*40)
print(f"Hello {name}!")
print(f"At {age}, you are a '{category}'.")
print(f"Your hobby is {hobby}. {hobby_message}")
print("Keep growing, keep glowing! ✨")
print("="*40)
```
