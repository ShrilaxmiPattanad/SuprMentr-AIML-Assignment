```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: Create dataset
data = {
    "Study_Hours": [1,2,3,4,5,6,7,8],
    "Marks": [35,40,50,55,65,70,80,90]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# Step 2: Identify features and labels
X = df[["Study_Hours"]]   # Feature
y = df["Marks"]           # Label

# Step 3: Train model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predict marks for new study hours
new_data = pd.DataFrame({"Study_Hours":[6]})
prediction = model.predict(new_data)

print("\nPredicted Marks for 6 Study Hours:", prediction[0])
```
