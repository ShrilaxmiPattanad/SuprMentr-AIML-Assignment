```python
# Spam Classifier using Machine Learning (Naive Bayes)

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Step 1: Create Email Dataset
emails = [
    "Win money now",
    "Important meeting tomorrow",
    "Claim your free gift",
    "Project discussion today",
    "Limited time discount offer",
    "Submit your assignment",
    "You won a lottery prize",
    "Team meeting at 3pm"
]

labels = [
    "Spam",
    "Not Spam",
    "Spam",
    "Not Spam",
    "Spam",
    "Not Spam",
    "Spam",
    "Not Spam"
]

# Convert to DataFrame
data = pd.DataFrame({
    "Email_Text": emails,
    "Category": labels
})

print("Dataset:\n")
print(data)

# Step 2: Convert text into numerical features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data["Email_Text"])

# Step 3: Target values
y = data["Category"]

# Step 4: Train Spam Classifier
model = MultinomialNB()
model.fit(X, y)

# Step 5: Test with new email
new_email = ["Congratulations you won free tickets"]

new_email_vector = vectorizer.transform(new_email)

prediction = model.predict(new_email_vector)

print("\nTest Email:", new_email[0])
print("Prediction:", prediction[0])
```
