```python
#Word Importance Explorer - 24-03-2026
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Sample documents
documents = [
    "Machine learning is a powerful technology",
    "Artificial intelligence and machine learning are related",
    "Deep learning is a part of machine learning",
    "Data science uses machine learning techniques",
    "AI and data science are trending fields"
]

# TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

# Convert to DataFrame
df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print("TF-IDF Matrix:\n")
print(df)

# Extract top keywords
print("\nTop Keywords per Document:")
for i, row in df.iterrows():
    top_words = row.sort_values(ascending=False).head(2)
    print(f"Document {i+1}: {list(top_words.index)}")
```
