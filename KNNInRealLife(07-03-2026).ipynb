```python
# Netflix-like Recommendation using KNN and Cosine Similarity

import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Step 1: Create movie rating dataset
data = {
    "User": ["User1", "User2", "User3", "User4", "User5"],
    "Action_Movie": [5, 4, 1, 2, 5],
    "Comedy_Movie": [3, 2, 5, 4, 3],
    "Drama_Movie": [4, 5, 1, 2, 4],
    "SciFi_Movie": [5, 4, 2, 1, 5]
}

df = pd.DataFrame(data)

print("User Movie Ratings Dataset:")
print(df)

# Step 2: Extract movie ratings
X = df.drop("User", axis=1)

# Step 3: Train KNN model using cosine similarity
knn = NearestNeighbors(n_neighbors=3, metric='cosine')
knn.fit(X)
# Step 4: Find users similar to User1
target_user = X.iloc[[0]]

distances, indices = knn.kneighbors(target_user)

print("\nUsers similar to User1:")
for i in indices[0][1:]:
    print(df.iloc[i]["User"])

# Step 5: Recommend movies liked by similar users
print("\nRecommended movies for User1 based on similar users:")
similar_users = df.iloc[indices[0][1:]]

avg_ratings = similar_users.drop("User", axis=1).mean()

for movie, rating in avg_ratings.items():
    if rating >= 4:
        print(movie)
```
