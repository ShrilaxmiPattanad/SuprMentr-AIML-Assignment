```python
# ==========================================================
# DATASET DETECTIVE
# Load dataset, explore data, find highest values,
# count missing values, and generate insights
# ==========================================================

import pandas as pd
import numpy as np

# Step 1: Create sample dataset
data = {
    "Name": ["Asha", "Rahul", "Priya", "Kiran", "Sneha", "Arjun", "Meena", "Vikram"],
    "Maths": [85, 78, 92, 70, 88, np.nan, 95, 60],
    "Science": [90, 82, 95, np.nan, 85, 80, 92, 65],
    "English": [88, 80, 94, 75, 90, 78, 96, np.nan],
    "Attendance": [92, 85, 98, 75, 90, 88, 99, 70]
}

df = pd.DataFrame(data)

# Step 2: Display top rows
print("Top 5 Rows of Dataset:\n")
print(df.head())

# Step 3: Find highest value in each column
print("\nHighest values in each numeric column:\n")
print(df.max(numeric_only=True))

# Step 4: Count missing values
print("\nMissing values count:\n")
print(df.isnull().sum())

# Step 5: Basic statistics
print("\nDataset Statistics:\n")
print(df.describe())

# Step 6: Generate insights
print("\nInsights:")
print("1. Highest Maths score is:", df["Maths"].max())
print("2. Highest Science score is:", df["Science"].max())
print("3. Highest English score is:", df["English"].max())
print("4. Average Attendance is:", df["Attendance"].mean())
print("5. Total missing values in dataset:", df.isnull().sum().sum())
```
