```python
# Movie reviews Analyzer - 26-03-2026
reviews = [
    "This movie was amazing and inspiring!",
    "Worst movie I have ever seen.",
    "It was okay, not great.",
    "Fantastic acting and story!",
    "Very boring and slow movie."
]

positive_words = ["amazing", "fantastic", "inspiring", "great", "good"]
negative_words = ["worst", "boring", "bad", "slow", "poor"]

def analyze_sentiment(review):
    review = review.lower()
    pos = sum(word in review for word in positive_words)
    neg = sum(word in review for word in negative_words)

    if pos > neg:
        return "Positive"
    elif neg > pos:
        return "Negative"
    else:
        return "Neutral"

# Output
for r in reviews:
    print(f"Review: {r}")
    print(f"Sentiment: {analyze_sentiment(r)}\n")
```
