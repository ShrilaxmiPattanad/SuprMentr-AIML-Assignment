```python
# ML IDEA GENERATOR DEMONSTRATION
# This program shows how Machine Learning works using Input -> Output

import pandas as pd
from sklearn.linear_model import LinearRegression

# ---------------------------------------------------
# 1. COLLEGE DOMAIN
# Problem: Predict student marks based on study hours
# ---------------------------------------------------

print("----- COLLEGE ML MODEL -----")

college_data = {
    "Study_Hours": [2,3,4,5,6,7],
    "Attendance": [60,65,70,75,80,85],
    "Marks": [50,55,60,65,70,80]
}

college_df = pd.DataFrame(college_data)

# Inputs (Features)
X_college = college_df[["Study_Hours","Attendance"]]

# Output (Target)
y_college = college_df["Marks"]

# Train the model
college_model = LinearRegression()
college_model.fit(X_college, y_college)

# New student data for prediction
new_student = pd.DataFrame({
    "Study_Hours":[6],
    "Attendance":[80]
})

college_prediction = college_model.predict(new_student)

print("Input (Study Hours=6, Attendance=80)")
print("Predicted Marks:", college_prediction[0])


# ---------------------------------------------------
# 2. HEALTHCARE DOMAIN
# Problem: Predict health risk using medical data
# ---------------------------------------------------

print("\n----- HEALTHCARE ML MODEL -----")

health_data = {
    "Age": [30,40,50,60,35,45],
    "Cholesterol": [180,200,220,250,190,210],
    "Risk_Score": [20,40,60,80,30,50]
}

health_df = pd.DataFrame(health_data)

# Inputs
X_health = health_df[["Age","Cholesterol"]]

# Output
y_health = health_df["Risk_Score"]

health_model = LinearRegression()
health_model.fit(X_health, y_health)

# New patient data
new_patient = pd.DataFrame({
    "Age":[50],
    "Cholesterol":[230]
})

health_prediction = health_model.predict(new_patient)

print("Input (Age=50, Cholesterol=230)")
print("Predicted Health Risk Score:", health_prediction[0])


# ---------------------------------------------------
# 3. SHOPPING DOMAIN
# Problem: Predict customer purchase amount
# ---------------------------------------------------

print("\n----- SHOPPING ML MODEL -----")

shopping_data = {
    "Visits": [1,2,3,4,5,6],
    "Time_Spent": [5,10,15,20,25,30],
    "Purchase_Amount": [100,200,300,400,500,600]
}

shopping_df = pd.DataFrame(shopping_data)

# Inputs
X_shop = shopping_df[["Visits","Time_Spent"]]

# Output
y_shop = shopping_df["Purchase_Amount"]

shopping_model = LinearRegression()
shopping_model.fit(X_shop, y_shop)

# New customer data
new_customer = pd.DataFrame({
    "Visits":[4],
    "Time_Spent":[20]
})

shop_prediction = shopping_model.predict(new_customer)

print("Input (Visits=4, Time Spent=20 minutes)")
print("Predicted Purchase Amount:", shop_prediction[0])

print("\nML Idea Demonstration Completed Successfully!")
```
