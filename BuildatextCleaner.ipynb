```python
# ===============================
# BUILD A TEXT CLEANER - 21-3-2026
# ===============================

!pip install nltk

import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download required resources
nltk.download('punkt')
nltk.download('punkt_tab')   # 🔥 FIX
nltk.download('stopwords')

# Sample text
text = "This is an Example!!! of TEXT cleaning, with punctuation & stopwords. It is very useful!!!"

print("🔹 Original Text:\n", text)

# Step 1: Lowercase
text_lower = text.lower()

# Step 2: Remove punctuation
text_no_punct = text_lower.translate(str.maketrans('', '', string.punctuation))

# Step 3: Tokenization
words = word_tokenize(text_no_punct)

# Step 4: Remove stopwords
filtered_words = [word for word in words if word not in stopwords.words('english')]

# Step 5: Final cleaned text
clean_text = " ".join(filtered_words)

print("\n🔹 Cleaned Text:\n", clean_text)
```
