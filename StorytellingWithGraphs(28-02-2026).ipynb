```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample dataset
data = {
    "Age": [22, 25, 47, 52, 46, 56, 48, 30, 28, 35],
    "Salary": [25000, 32000, 52000, 58000, 50000, 60000, 55000, 42000, 39000, 45000],
    "Experience": [1, 3, 10, 12, 9, 14, 11, 5, 4, 6],
    "Purchased": [0, 0, 1, 1, 1, 1, 1, 0, 0, 1]
}

df = pd.DataFrame(data)

# --------------------------------
# 1 Histogram (Data Distribution)
# --------------------------------
plt.figure()
plt.hist(df["Salary"], bins=5)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

# --------------------------------
# 2 Bar Chart (Category Comparison)
# --------------------------------
plt.figure()
df["Purchased"].value_counts().plot(kind="bar")
plt.title("Purchased vs Not Purchased")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# --------------------------------
# 3 Scatter Plot (Feature Relation)
# --------------------------------
plt.figure()
plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

# --------------------------------
# 4 Line Plot (Trend)
# --------------------------------
plt.figure()
plt.plot(df["Experience"], df["Salary"])
plt.title("Experience vs Salary Trend")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()

# --------------------------------
# 5 Box Plot (Outlier Detection)
# --------------------------------
plt.figure()
sns.boxplot(x=df["Salary"])
plt.title("Salary Boxplot")
plt.show()

# --------------------------------
# 6 Heatmap (Correlation Matrix)
# --------------------------------
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()

# --------------------------------
# 7 Pair Plot (Feature Relationships)
# --------------------------------
sns.pairplot(df)
plt.show()

# --------------------------------
# 8 Confusion Matrix Example
# --------------------------------
from sklearn.metrics import confusion_matrix
import seaborn as sns

y_true = [0,0,1,1,1,1,0,0,1,1]
y_pred = [0,1,1,1,0,1,0,0,1,1]

cm = confusion_matrix(y_true, y_pred)

plt.figure()
sns.heatmap(cm, annot=True, fmt="d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
