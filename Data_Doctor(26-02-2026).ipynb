```python
# Data Doctor Assignment - Advanced Dataset Cleaning

import pandas as pd
import numpy as np

# -----------------------------
# Step 1: Create Sample Dataset
# -----------------------------
data = {
    "Name": ["Amit", "Sara", "John", "Sara", "Anu", "Ravi", "amit"],
    "Age": [20, 21, np.nan, 21, 22, 200, 20],
    "Department": ["CS", "IT", "CS", "IT", None, "cs", "CS"],
    "Marks": [85, 90, 78, 90, np.nan, 150, 85]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

# -----------------------------
# Step 2: Check Missing Values
# -----------------------------
print("\nMissing Values:")
print(df.isnull().sum())

# -----------------------------
# Step 3: Handle Missing Values
# -----------------------------
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df["Department"].fillna("Unknown", inplace=True)

# -----------------------------
# Step 4: Remove Duplicate Rows
# -----------------------------
df = df.drop_duplicates()

# -----------------------------
# Step 5: Standardize Text Data
# -----------------------------
df["Name"] = df["Name"].str.lower().str.strip()
df["Department"] = df["Department"].str.lower().str.strip()

# -----------------------------
# Step 6: Fix Outliers
# -----------------------------
# Remove unrealistic ages
df.loc[df["Age"] > 100, "Age"] = df["Age"].median()

# Remove unrealistic marks
df.loc[df["Marks"] > 100, "Marks"] = df["Marks"].mean()

# -----------------------------
# Step 7: Data Summary Report
# -----------------------------
print("\nCleaned Dataset:\n")
print(df)

print("\nDataset Summary:")
print(df.describe())

print("\nUnique Departments:", df["Department"].unique())

# -----------------------------
# Step 8: Save Clean Dataset
# -----------------------------
df.to_csv("cleaned_students.csv", index=False)

print("\nData cleaning completed. File saved as 'cleaned_students.csv'")
```
