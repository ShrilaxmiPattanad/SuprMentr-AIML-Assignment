```python
# ===============================
# TEXT CHALLENGES - 20-3-2026
# ===============================

# Import libraries
import re

# Step 1: Define messy sentences
sentences = [
    "I luv this movie!!! 😍🔥",
    "Hiii broooo how r u???",
    "This is awsm!!!",
    "LOL that was funny 😂",
    "gud mrng frnds",
    "I can’ttt believe this!!!",
    "OMG!!! this is lit 🔥",
    "Plzz help meee!!!",
    "Thx for ur support",
    "Wat r u doing??",
    "Soooo happpyyy!!!",
    "That’s gr8 news",
    "U r amazing 😎",
    "C u soon!!!",
    "Dis is not gud",
    "I’m soooo tired 😴",
    "Hehe that was funnn",
    "Best day everrrr!!!",
    "Yessss finally!!!",
    "I’m loving ittt 😍"
]

# Step 2: Slang dictionary
slang_dict = {
    "luv": "love", "r": "are", "u": "you", "lol": "laughing",
    "gud": "good", "mrng": "morning", "frnds": "friends",
    "omg": "oh my god", "plzz": "please", "thx": "thanks",
    "wat": "what", "gr8": "great", "c": "see", "dis": "this"
}

# Step 3: Detect emojis/symbols
def detect_emojis(text):
    return re.findall(r'[^\w\s]', text)

# Step 4: Detect repeated characters
def detect_typos(text):
    return re.findall(r'(.)\1{2,}', text)

# Step 5: Clean text function
def clean_text(text):
    text = text.lower()  # lowercase

    # Remove emojis & punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Replace slang words
    words = text.split()
    words = [slang_dict[word] if word in slang_dict else word for word in words]

    # Remove repeated characters
    words = [re.sub(r'(.)\1{2,}', r'\1\1', word) for word in words]

    return " ".join(words)

# Step 6: Process and display results
for i, sentence in enumerate(sentences, 1):
    print(f"\n🔹 Sentence {i}: {sentence}")

    emojis = detect_emojis(sentence)
    typos = detect_typos(sentence)

    print("👉 Detected Emojis/Symbols:", emojis)
    print("👉 Detected Repeated Characters:", typos)

    cleaned = clean_text(sentence)
    print("✅ Cleaned Text:", cleaned)
```
