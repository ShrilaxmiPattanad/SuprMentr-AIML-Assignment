```python
# ==========================================================
# NUMPY SPEED TEST: Python List vs NumPy Array
# Compare execution time with 1 Million numbers
# ==========================================================

import time
import numpy as np

# Step 1: Create Python list and NumPy array with 1 Million numbers
python_list = list(range(1000000))
numpy_array = np.arange(1000000)

# ----------------------------------------------------------
# Step 2: Test Python List Speed (Addition operation)
# ----------------------------------------------------------

start_time = time.time()

python_result = [x + 5 for x in python_list]

end_time = time.time()

python_time = end_time - start_time

print("Python List Execution Time:", python_time, "seconds")

# ----------------------------------------------------------
# Step 3: Test NumPy Array Speed (Addition operation)
# ----------------------------------------------------------

start_time = time.time()

numpy_result = numpy_array + 5

end_time = time.time()

numpy_time = end_time - start_time

print("NumPy Array Execution Time:", numpy_time, "seconds")

# ----------------------------------------------------------
# Step 4: Compare Results
# ----------------------------------------------------------

print("\nSpeed Difference:")
print("NumPy is", python_time / numpy_time, "times faster than Python List")

# ----------------------------------------------------------
# Step 5: Observations
# ----------------------------------------------------------

print("\nObservations:")
print("1. NumPy arrays execute faster than Python lists.")
print("2. NumPy uses optimized vectorized operations.")
print("3. Python lists take more time due to element-wise processing.")
```
